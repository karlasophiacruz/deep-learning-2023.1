#### NumPy  Fibonacci Series using Binet Formula

image.png

```python
def fibonacci(n):
  a = np.arange(1, n + 1)

  sqrt5 = np.sqrt(5)

  alpha = (1 + sqrt5) / 2
  beta = (1 - sqrt5) / 2

  fn = np.rint( ( (alpha ** a) - (beta ** a) ) / sqrt5)

  fn = np.insert(fn, 0, 0)

  return fn

n = 10
print("The first {} numbers of Fibonacci series are {}. " .format(n, fibonacci(n)))
```
